### 굳이 스크롤을 내려야 함 
- 스크롤을 내리면 html이 변경 됨 -> 동적인 페이지
- 그럼 html 다시 호출해야 함

```python
# n번만 스크롤 내리기
def scroll_n_times(driver, n, interval=0.02):
    body = driver.find_element("tag name", "body")
    for _ in range(n):
        body.send_keys("\ue00f")  # PAGE_DOWN
        time.sleep(interval)

# 사용 예시
scroll_n_times(driver, n=30)
```

# 명령
- 데이터프레임 book에 d_url이 실제 각 페이지의 주소이다.
- 남성에서 가장 높은 %를 보이는 연령대, 여성에서 가장 높은 %를 보이는 연령대를 추출해서 
'남성_top_연령대', '여성_top_연령대'에 추가.
- 첫 3개 페이지만 테스트 해보고 싶다.
- 패키지 로드와 드라이버 세팅 코드 필요없음.

### 데이터프레임에 직접 적용해보기 위한 코드를 챗봇으로부터 추출(첫 3개 행만)

```python
driver = webdriver.Chrome(options=options)
driver.get(url)

# 결과 저장용 컬럼 초기화
book['남성_top_연령대'] = None
book['여성_top_연령대'] = None

# 앞 3개 페이지만 테스트
for idx in book.index[:3]:
    url = book.loc[idx, 'd_url']
    driver.get(url)

    scroll_n_times(driver, n=30)

    soup = BeautifulSoup(driver.page_source, 'html.parser')

    buyer_area = soup.find('div', class_='buyer_area_a')
    if buyer_area is None:
        continue

    rows = buyer_area.find_all('tr')

    male_data = {}
    female_data = {}

    for row in rows:
        age_td = row.find('td', class_='tb_tit')
        woman_td = row.find('td', class_='tb_woman')
        man_td = row.find('td', class_='tb_man')

        if not (age_td and woman_td and man_td):
            continue

        age = age_td.get_text(strip=True)

        woman_per = woman_td.find('div', class_='per')
        man_per = man_td.find('div', class_='per')

        if woman_per and man_per:
            female_data[age] = float(woman_per.get_text(strip=True).replace('%', ''))
            male_data[age] = float(man_per.get_text(strip=True).replace('%', ''))

    # 최댓값 연령대 추출
    if male_data:
        book.loc[idx, '남성_top_연령대'] = max(male_data, key=male_data.get)

    if female_data:
        book.loc[idx, '여성_top_연령대'] = max(female_data, key=female_data.get)
```
