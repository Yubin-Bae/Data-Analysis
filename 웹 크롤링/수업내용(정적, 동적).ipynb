### 웹 크롤링

### 정적
1. requests로 target url에 html정보 요청
2. bs4 패키지로 해석
3. 원하는 데이터 추출(태그와 find, find_all)

### 동적
1. selenium 패키지로 target 소스에 접근
    - 예) 마우스로 클릭, 로그인, 스크롤내리기, 새탭에서, 페이지 안에 위젯
    (iframe)

2. 'driver'에서 html 요청
3. bs4 패키지로 해석
4. 원하는 데이터 추출(태그와 find, find_all)
