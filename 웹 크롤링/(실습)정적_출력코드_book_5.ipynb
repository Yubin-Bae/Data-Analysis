```python
### 패키지 정리 및 html파싱
import requests
from bs4 import BeautifulSoup
import pandas as pd

# static web html (정적 웹)
url = 'https://www.aladin.co.kr/shop/wproduct.aspx?ItemId=333420348'   
html = requests.get(url)                                                
soup = BeautifulSoup(html.text, 'html.parser')

# Sales Point
sp = soup.find('div', 'Ere_fs15 Ere_ht18')
세일즈포인트 = sp.find_all('strong')[0].text

# 가격
가격 = soup.find('em', 'Ere_fs24 samnung_pink').text

# 쪽수
info_1 = soup.find('div', 'conts_info_list1')
for i in info_1.find_all('li'):
    if i.text.endswith("쪽"):
        쪽수 = i.text[:-1]
print("쪽수:", 쪽수)
print("=" * 32)

# 3분류
info_2 = soup.find('div', 'conts_info_list2')
분류_라인 = info_2.find_all('li')[0]
분류 = 분류_라인.find_all('a')[2]
분류 = 분류.text

# 리뷰 수
review = soup.find('div', 'info_list Ere_fs15 Ere_ht18')
리뷰 = review.find_all('a')[3].text.replace('리뷰(','').replace(')','')

# 평균별점
별점 = soup.find('a', 'Ere_sub_pink Ere_fs16 Ere_str').text.strip()

# 결과출력
print("세일즈포인트|가격|쪽수|분류|리뷰|별점")                 # 보기 편하라고 추가함
print(f'{세일즈포인트}|{가격}|{쪽수}|{분류}|{리뷰}|{별점}')
```
