### 입력값: book_test 데이터프레임에 'isbn13'열 입니다.
### 반환값: 위 코드의 결과입니다. (상세페이지로 가는 url)
### 반환값이 저장될 장소? -> 새로운 열 'd_url' (2차 웹 크롤링 용 url)
### 위 코드를 함수로 만들고 (에러 제어 없이) 데이터프레임 book_test에 적용하는 코드 작성
### 추가정보 없이 코드만 출력

```python
import requests
from bs4 import BeautifulSoup

def get_detail_url(isbn13):
    url = f"https://www.aladin.co.kr/search/wsearchresult.aspx?SearchWord={isbn13}"
    html = requests.get(url)
    soup = BeautifulSoup(html.text, 'html.parser')
    return soup.find('a', 'bo3')['href']

book_test['d_url'] = book_test['isbn13'].apply(get_detail_url)
```
