```python
import requests
from bs4 import BeautifulSoup

url = f'https://search.naver.com/search.naver?query=%EB%84%A4%EC%9D%B4%EB%B2%84+%EC%9D%8C%EC%9B%90%EC%B0%A8%ED%8A%B8'
html = requests.get(url)
soup = BeautifulSoup(html.text, 'html.parser')

```

```python
soup
```

```python
# 데이터 찾기
# 랭킹
print("<네이버 음원차트 TOP 10>")

랭킹html = soup.find('div', 'chart_list_wrap _sap_list')
랭킹html.text

# 아티스트
아티스트html = soup.find_all("span", "name")

아티스트list = []
for artist in 아티스트html:
    아티스트list.append(artist.text)    
아티스트list[:10]

# 타이틀
타이틀html = soup.find_all("span", "txt" )

타이틀list = []
for txt in 타이틀html:
    타이틀list.append(txt.text)
타이틀list[:10]

# 앨범명
앨범명html = soup.find_all("a", "album")

앨범명list = []
for album in 앨범명html:
    앨범명list.append(album.text)
앨범명list[:10]

# 출력
print("순위\t타이틀\t아티스트\t앨범명")
for i in range(10):
    print(f"{i+1}\t{타이틀list[i]}\t{아티스트list[i]}\t{앨범명list[i]}")    
print("\n")
```

```python
import pandas as pd
data = {'타이틀': 타이틀list[:10], '아티스트': 아티스트list[:10], '앨범명': 앨범명list[:10]}
df = pd.DataFrame(data)
df.index = df.index + 1     # 인덱스를 1부터 시작
df
```
