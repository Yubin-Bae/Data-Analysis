```python
import requests
from bs4 import BeautifulSoup

def naver_weather():
    city = input("지역을 입력하세요(예:종로구): ")
    url = f"https://search.naver.com/search.naver?query={city}+날씨"

    headers = {'User-Agent': 'Mozilla/5.0'}
    res = requests.get(url, headers=headers)
    soup = BeautifulSoup(res.text, 'html.parser')

    온도html = soup.find('div', class_='temperature_text')
    온도 = 온도html.text.strip().replace('현재 온도', '').replace('°','')

    상태 = soup.find('span', class_='weather before_slash').get_text(strip=True)

    items = soup.find_all('div', class_='sort')
    체감온도 = items[0].text.strip().replace('체감', '').replace('°','')
    습도 = items[1].text.strip().replace('습도', '').replace('%','')

    print("지역:", city)
    return f'온도:{온도}, 상태:{상태}, 체감온도:{체감온도}, 습도:{습도}'

naver_weather()
```
