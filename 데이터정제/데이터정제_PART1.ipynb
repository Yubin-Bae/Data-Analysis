# 중복처리
- 해야됨
- 열에 따라 중복을 꼭 확인해야 되는 경우가 있다.
(상품ID, 식별자, 주민등록번호, 등등)
  * 고유값 : isbn13, 도서명
  * 범주형, 연속형 : 중복이 당연하다.

## duplicated()
  - PART1 기본 -> 중복값: T, 최초 등장: F
    * duplicated() : 최초값 뻬고 나머지 중복만 나온다.
     
      => [주로 중복데이터 갯수 확인]

    * duplicated()(keep=False) : 최초값 포함 중복이 나온다. (중복값이면 모두 True를 반환)
      
      => [중복 데이터끼리 묶어서 출력 후 확인용]


---


  
  - PART2 반전 -> 실제 처리할 때 사용, 유지 대상: T, 제거 대상: F
    * ~duplicated() : 최초값, 유일값만 남긴다. (중복X)

      => [주로 처리용]
      
    * ~duplicated(keep=False) 원래부터 1개였던(유일값) 만 남긴다.


---



* duplicated() : 중복된 행을 찾을 때 -> 제거 대상 식별
* ~duplicated() : 유지할 행을 찾을 때 -> 정제 후 데이터 선택

```python
df_sample = pd.DataFrame({
        'id':['a','a','b','c','b'],
        'age':[20, 30, 40, 50, 50]
})
df_sample
```

```python
# .duplicated()
df_sample['id'].duplicated()
```

```python
# .duplicated() 반전 => 실제 처리할 때 사용
~df_sample['id'].duplicated()
```

```python
# 단순하게 True 개수 셈
# 중복을 센거라면 최초값을 제외한 중복개수
sum(df_sample['id'].duplicated())
```

```python
sum(df_sample['id'].duplicated())
sum(book2['isbn13'].duplicated())
```

```python
# 유일한 값만 False, 중복값 모두 True
df_sample['id'].duplicated(keep=False)
```

```python
# (반전) 유일한 값만 True, 중복값 모두 False
~df_sample['id'].duplicated(keep=False)
```

#**GroupBy**
- 데이터를 그룹화하고 그룹화된 데이터에 대한 연산을 수행하는 기능
- 내가 원하는 열로 인덱스를 새로 지정할 수 있음.
- **[df.groupby(by='열이름)['보고싶은 열'].count()**


```python
data = {'이름': ['김철수', '박영희', '김철수', '박영희', '이민수'],
        '과목': ['수학', '수학', '영어', '영어', '수학'],
        '점수': [85, 95, 78, 92, 88]}
score = pd.DataFrame(data)
score
```

```python
# 학생별 전과목 평균 점수
score.groupby('이름')['점수'].mean()  # 평균
#                              max() : 최대값
#                             count(): 빈도
#                             median() : 중앙값
#                               std() : 표준편차
```

```python
# 과목별 최대점수
score.groupby('과목')['점수'].max()
```

```python
# 학생별로 각 과목의 점수
# 열이 2개일 경우 리스트로 묶음
score.groupby(['이름', '과목'])['점수'].mean()
```

```python
# 출판사별 대출건수 최대값
book3.groupby('publisher')['loan_count'].max()
```

```python
# 출판사별, 분류별, 리뷰수의 평균
book3.groupby(['publisher', '분류'])['review_count'].mean()
```

```python
# 분류별 페이지수의 평균
book3.groupby('분류')['page_count'].mean()
```

## **groupby**
- groupby로 isbn13이 같은도서의 대출건수를 합침

```python
isbngroup = book3.groupby('isbn13')['loan_count'].sum()
```

```python
# isbngroup의 인덱스는 isbn
# isbn 기준으로 대출건수 합 확인 완료
isbngroup[9788901033518]
```

- mapping -> 맵핑테이블을 통해 원본 df에 키를 찾아 값을 추가하거나 바꿈
- isbngroup 결과물은 키(인덱스)가 isbn13이고 대출건수가 연산이 되어 있다. 즉, 맵핑테이블로 사용.

```python
# 대출건수 바뀌었는지 확인 (loan_count)
book3['loan_count'] = book3['isbn13'].map(isbngroup)
book3[book3['isbn13'] == 9788901033518]
```

## mapping을 통해 대출건수에 새로운 값으로 교체

## 중복제거
- 대출건수가 업데이트 되었기 때문에 최초값을 제외한 중복은 모두 제거한다.

```python
book4 = book3[~book3['isbn13'].duplicated()]
```

```python
book4.info()
```

## 도서명 중복처리용 샘플

```python
# isbn 중복은 이제 없음
book4[book4['bookname'].duplicated(keep=False)].sort_values('bookname')
```

- (간니닌니) 마법의 도서관 :명작 속으로 떠나는 판타지 동화 여행(472, 420 등등)
- 흰 :한강 소설(1463, 2562)

```python
# 2번 실행 X => 평균 잘못 구해짐
# 처리
## 맵핑테이블 생성
namegroup = book4.groupby('bookname')['loan_count'].mean()
#                                                   대출건수의 평균

## 맵핑
book4['loan_count'] = book4['bookname'].map(namegroup)

## 중복제거
book4 = book4[~book4['bookname'].duplicated()]

## 확인
book4[book4['bookname'] == '흰 :한강 소설 ']
```
