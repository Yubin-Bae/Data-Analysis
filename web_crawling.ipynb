- 추출된 모든 상품을 데이터프레임(이름은 musinsa)으로 만드는 코드
- 추가정보 없이 코드만 작성

```python
import pandas as pd
from bs4 import BeautifulSoup

# 페이지 소스 파싱
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

data = []

# 각 상품 카드 기준으로 순회
items = soup.find_all("div", class_="UIProductColumn__Wrap-sc-1t5ihy5-0")

for item in items:
    # 제품명
    name_tag = item.find(
        "p",
        class_="text-body_13px_reg line-clamp-2 break-all whitespace-break-spaces text-black font-pretendard"
    )
    name = name_tag.get_text(strip=True) if name_tag else None

    # 가격
    price_tag = item.find(
        "span",
        class_="text-body_13px_semi UIProductColumn__PriceText-sc-1t5ihy5-11 cZhItm text-black font-pretendard"
    )
    price = price_tag.get_text(strip=True) if price_tag else None

    # 제품 ID
    product_id = item.get("data-item-id")

    # 제품 링크
    link_tag = item.find(
        "a",
        class_="UIProductColumn__Anchor-sc-1t5ihy5-2 jAWGzh gtm-select-item"
    )
    link = link_tag.get("href") if link_tag else None

    data.append({
        "product_name": name,
        "price": price,
        "product_id": product_id,
        "product_link": link
    })

# 데이터프레임 생성
musinsa = pd.DataFrame(data)
```

## class_를 붙이는 경우
- HTML 속성 이름을 지정할 때만 붙임. 그 외에는 x
ex) BeautifulSoup의 find/find_all에서 *HTML 속성을 지정할 때*
<span class="Category__CategoryList-sc-x0nx70-2 kavYEX">

- class_를 안붙이는 경우
- 변수 이름 => 분류html = ...
- 인덱싱 => soup.find_all(...)[1]
- .text, .get_text() => tag.text


```python
import pandas as pd

# 분류
분류html = soup.find_all('span', class_="Category__CategoryList-sc-x0nx70-2 kavYEX")[1]
분류 = 분류html.get_text(strip=True) if 분류html else None                         # 분류html이 비어 있지 않으면 A 실행, 비어 있으면 None 반환

# 브랜드
브랜드html = soup.find('span', class_='text-body_14px_med font-pretendard')
브랜드 = 브랜드html.get_text(strip=True) if 브랜드html else None                    # 브랜드html이 비어 있지 않으면 A 실행, 비어 있으면 None 반환

# 첫 번째 리뷰 - 사용자 닉네임
닉네임html = soup.find('span', class_='text-body_13px_med UserProfileSection__Nickname-sc-1aa5han-4 cxOAeE text-black font-pretendard')
닉네임 = 닉네임html.get_text(strip=True) if 닉네임html else None                    # 닉네임html이 비어 있지 않으면 A 실행, 비어 있으면 None 반환

# 리뷰내용
리뷰html = soup.find('div', class_='ExpandableContent__ContentContainer-sc-1951ro-1 igSvLL')
리뷰내용 = 리뷰html.get_text(strip=True) if 리뷰html else None                      # 리뷰html이 비어 있지 않으면 A 실행, 비어 있으면 None 반환

musinsa = pd.DataFrame({
    '분류': [분류],     
    '브랜드': [브랜드],
    '닉네임': [닉네임],
    '리뷰내용': [리뷰내용]
})

musinsa
```
