## 필요한 데이터
- span태그 class = "txt" -> 곡 제목
- span태그 class = "name" -> 가수명
- a태크 class = "album" -> 앨범명

## html정보
- 첨부파일 확인

```python
import requests
from bs4 import BeautifulSoup

url = 'https://search.naver.com/search.naver?query=음원차트'
html = requests.get(url)
soup = BeautifulSoup(html.text, 'html.parser')
box = soup.find('div', 'chart_list_wrap _sap_list')

titles = box.find_all('span', class_='txt')
artists = box.find_all('span', class_='name')
albums = box.find_all('a', class_='album')

result = []

for title, artist, album in zip(titles, artists, albums):
    result.append({
        '곡명': title.get_text(strip=True),
        '가수': artist.get_text(strip=True),
        '앨범': album.get_text(strip=True)
    })

result
```

```python
import pandas as pd
df = pd.DataFrame(result)
df
```
